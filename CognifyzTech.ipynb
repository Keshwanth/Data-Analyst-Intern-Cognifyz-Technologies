```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Viewing the dataset
data=pd.read_csv("/content/drive/MyDrive/Dataset .csv")
data.head(),data.columns
```

# **LEVEL 1**

#TASK 1
Task: Top Cuisines

Determine the top three most
common cuisines in the dataset.
Calculate the percentage of
restaurants that serve each of the top
cuisines.


```python
#Finding the top 3 most common cuisines
cuisines_series=data['Cuisines'].str.split(', ').explode().value_counts()
top_cuisines=cuisines_series.head(3)
print("The top 3 Cuisines are\n",top_cuisines)

#Calculating % of restaurants that serve the top cuisines
total_restaurants=len(data)
top_cuisines_percentage=(top_cuisines/total_restaurants)*100
print("\nPercentage of restaurants that serve top cuisines are\n",top_cuisines_percentage)
```

#TASK 2

Task: City Analysis

Identify the city with the highest number
of restaurants in the dataset.
Calculate the average rating for
restaurants in each city.
Determine the city with the highest
average rating.

```python
#Finding City with High No. of restaurants
city_most_restaurants=data['City'].value_counts().idxmax()
print("City with high restaurants :",city_most_restaurants)

#Calculating the average rating for restaurants in each city
average_rating_by_city=data.groupby('City')['Aggregate rating'].mean()
print("\nAverage Rating by each city\n",average_rating_by_city)

#Finding the city with the highest average rating
city_highest_average_rating=average_rating_by_city.idxmax()

print("\nCity with Highest Average Rating :",city_highest_average_rating,"(",average_rating_by_city[city_highest_average_rating],")")
print()
```

#TASK 3
Task: Price Range Distribution

Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.
Calculate the percentage of restaurants
in each price range category.


```python
#Creating a bar chart to visualize the distribution of price ranges among restaurents
price_range=data['Price range'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
price_range.plot(kind='bar',color='skyblue')
plt.title('Price Range Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('No. of Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()

# Calculate the percentage of restaurants in each price range category
total_restaurants=len(data)
price_range_percentage=(price_range/total_restaurants)*100
print("\nPercentage of Restaurants in each price range category :\n",price_range_percentage)
```

#TASK 3
Task: Geographic Analysis

Plot the locations of restaurants on a
map using longitude and latitude
coordinates.
Identify any patterns or clusters of
restaurants in specific areas.


```python
#Ploting the locations of restaurants on a graph using longitude and latitude coordinates as x & y axis
plt.figure(figsize=(10, 6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.6, c='blue', edgecolors='w', s=50)
plt.title('Scatter Plot of Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
#Identify any clusters of restaurants in specific areas.
from sklearn.cluster import KMeans
coordinates = data[['Longitude','Latitude']]
kmeans=KMeans(n_clusters=5)
clusters=kmeans.fit_predict(coordinates)
data['Cluster'] = clusters

plt.scatter(data['Longitude'], data['Latitude'], c=data['Cluster'],cmap='viridis',alpha=0.6,edgecolors='w')
plt.title('Clustered Restaurants by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

#TASK 4
Task: Restaurant Chains

Identify if there are any restaurant chains
present in the dataset.
Analyze the ratings and popularity of
different restaurant chains.

```python
#Identifying if there are any restaurant chains present in the dataset
restaurant_name_counts=data['Restaurant Name'].value_counts()
restaurant_chains=restaurant_name_counts[restaurant_name_counts>1] #Taking the the count of repeated restaurant chains
print(restaurant_chains.head(10))
print(f"Total number of restaurant chains identified: {len(restaurant_chains)}")

#Calculating the average rating and total votes for each chain
chain_data = data[data['Restaurant Name'].isin(restaurant_chains.index)]
chain_average_ratings = chain_data.groupby('Restaurant Name')['Aggregate rating'].mean()
chain_total_votes = chain_data.groupby('Restaurant Name')['Votes'].sum()
chain_analysis = pd.DataFrame({'Average Rating': chain_average_ratings,
                               'Total Votes': chain_total_votes
                               }).sort_values(by='Total Votes', ascending=False)
print("\n",chain_analysis.head(10))
```

#TASK 1

Task: Restaurant Reviews

Analyze the text reviews to identify the most
common positive and negative keywords.
Calculate the average length of reviews and
explore if there is a relationship between
review length and rating.

```python
import nltk
nltk.download('stopwords')
```

```python
#Analyzing the text reviews to identify the most common positive and negative keywords by count of each words repeated
from collections import Counter
from nltk.corpus import stopwords
import re

data['cleaned_reviews']=data['Rating text'].str.lower().apply(lambda x: re.sub('[^\w\s]', '', str(x)))
stop = stopwords.words('english')
data['cleaned_reviews'] = data['cleaned_reviews'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
word_counts=Counter(" ".join(data['cleaned_reviews']).split())
counts={word: count for word, count in word_counts.items()}
print("Keywords Count:",counts)
```

```python
#Calculating the average length of reviews to find if there is a relationship between review length and rating.
data['review_length']=data['cleaned_reviews'].str.len()
average_length=data['review_length'].mean()
print("Average Review Length:", average_length)

plt.scatter(data['Rating text'], data['review_length'], alpha=0.5)
plt.title('Review Length vs Rating')
plt.xlabel('Rating')
plt.ylabel('Review Length')
print("\n")
plt.show()
```

#TASK 2

Task: Votes Analysis

Identify the restaurants with the highest and
lowest number of votes.
Analyze if there is a correlation between the
number of votes and the rating of a
restaurant.

```python
#Identifying the restaurants with the highest and lowest number of votes.
highest_votes = data[data['Votes'] == data['Votes'].max()]
lowest_votes = data[data['Votes'] == data['Votes'].min()]
print("Restaurants with the Highest Number of Votes:","\n",highest_votes[['Restaurant Name', 'Votes']])
print("\nRestaurants with the Lowest Number of Votes:","\n",lowest_votes[['Restaurant Name', 'Votes']])
```

```python
#Analyzing if there is a correlation between the number of votes and the rating of a restaurant.
plt.figure(figsize=(10, 5))
plt.scatter(data['Votes'], data['Aggregate rating'], alpha=0.5)
plt.title('Correlation between No. of Votes and Ratings')
plt.xlabel('No. of Votes')
plt.ylabel('Ratings')
plt.show()

#Calculating the value ofcorrelation coefficient
correlation = data['Votes'].corr(data['Aggregate rating'])
print("\n","Correlation Coefficient between No. of Votes and Ratings:", correlation)
```

#TASK 3
Task: Price Range vs. Online Delivery and
Table Booking

Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.
Determine if higher-priced restaurants are
more likely to offer these services

```python
#Analyzing if there is a relationship between the price range and the availability of online delivery and table booking.
data = pd.read_csv('/Dataset .csv')
data['Has Online delivery'] = (data['Has Online delivery'] == 'Yes').astype(int)
data['Has Table booking'] = (data['Has Table booking'] == 'Yes').astype(int)
grouped_data = data.groupby('Price range').agg({
    'Has Online delivery': 'mean',
    'Has Table booking': 'mean'
}).reset_index()

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

ax[0].bar(grouped_data['Price range'], grouped_data['Has Online delivery'], color='blue')
ax[0].set_title('Price Range vs Online Delivery')
ax[0].set_xlabel('Price Range')
ax[0].set_ylabel('Proportion of Online Delivery Available')

ax[1].bar(grouped_data['Price range'], grouped_data['Has Table booking'], color='green')
ax[1].set_title('Price Range vs Table Booking')
ax[1].set_xlabel('Price Range')
ax[1].set_ylabel('Proportion of Table Booking Available')

plt.show()
```

```python
#Determining if higher-priced restaurants are more likely to offer these services
if not grouped_data.empty:
    for service in ['Has Online delivery', 'Has Table booking']:
        trend = grouped_data.corr().loc['Price range', service]
        print(f"The correlation between price range and {service.lower()} is {trend:.2f}. This indicates",
              "a positive trend" if trend > 0 else "a negative trend" if trend < 0 else "no clear trend")
else:
    print("No data available for analysis.")
```
